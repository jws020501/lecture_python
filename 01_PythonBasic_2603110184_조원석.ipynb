### 02.대입연산자
* 두 변수의 값을 교환하는 프로그램(Swap)

```python
a = 10
b = 20
print("a:",a,"b:",b)
temp = a 
a = b 
b = temp
print("a:",a,"b:",b)
```

```python
a = b = c = 7
print(a,b,c)
```

```python
a , b , c = 7 , 11, 19
print(a,b,c)
```

```python
a , b = 15 ,25
print(a,b)
a,b = b,a
print(a,b)
```

```python
a = 10>20
b = 10<20
print(a)
print(b)
```
