```python
#while문 연습 : 입력값에 따라 반복 종료 하기
#사용자로부터 숫자를 하나씩 입력 받아서 총합과 평균을 출력
#사용자가  0을 입력하면 숫자입력 멈춤
sum = count = 0
num = int(input("숫자입력(0은 끝): "))
while num != 0:
    sum += num
    count += 1
    num = int(input("숫자입력(0은 끝): "))

#0 으로 나누기를 하면 제로디비전에러 발생 -> 예외처리
if count !=0:
    average = sum/count
else:
    average = 0
print(f"sum: {sum}, average: {average}")
```
