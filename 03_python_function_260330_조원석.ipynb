```python
# 구구단 출력 함수
def gugudan(num):
    print(f"[{num}단]")
    for i in range(1,10):
        print(f"{num} x {i} = {num * i}")
```

```python
number =int(input())
gugudan(number)
```

```python
for i in range(1,10):
    gugudan(i)
    print("")
```

```python
#result 값 확인
def add_and_mul(a,b):
    return a+b, a*b

result = add_and_mul(3,4)
print(result)

result1 , result2 = add_and_mul(3,4)
print(result1 , result2)
```

```python
#지역변수 , 전역변수
sum = 10
def test(sum , num):
    sum += num
    return sum
sum = test(sum,5)
print(sum)
```

```python
#지역변수 , 전역변수 *기피할것
sum = 10
def test(num):
    global sum
    sum += num
test(5)
print(sum)
```
