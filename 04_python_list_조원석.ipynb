```python
# getIndex
number_list = [23,45,27,11,25,65,78]
def getIndex(l,n):
    return l.index(n)
print(getIndex(number_list,25))
```

```python
# getMax
number_list = [23,45,27,11,25,65,78]
def getMax(l):
    return max(l)
print(getMax(number_list))
```

```python
# getMin
number_list = [23,45,27,11,25,65,78]
def getMin(l):
    return min(l)
print(getMin(number_list))
```

```python
#countGT
number_list = [23,45,27,11,25,65,78]
def countGT(l,n):
    return sum(map(lambda x: x > n, l))
print(countGT(number_list,25))
```

```python
#sumList
number_list = [23,45,27,11,25,65,78]
def sumList(l):
    return sum(l)
print(sumList(number_list))
```

```python
#swapList
number_list = [23,45,27,11,25,65,78]
def swapList(l):
    return list(reversed(l))
print(swapList(number_list))
```

```python
def recursion(s, l, r, c):
    if l >= r: return [1,c] 
    elif s[l] != s[r]: return [0,c]
    else: return recursion(s, l+1, r-1,c+1)

def isPalindrome(s):
    return recursion(s, 0, len(s)-1,1)

n = int(input())
for _ in range(n):
    a = input()
    res = isPalindrome(a)
    print(f"{res[0]} {res[1]}")
```
