```python
##### 3개의 숫자중 최대값 구하기
# a, b, c = input().split()
# a, b, c = int(a),int(b),int(c)

a, b, c = map(int,input("세개의 수를 입력하시오").split())
max = a
if(b>max):
    max = b
if(c>max):
    max = c
print(f'가장큰 수는 {max}입니다')
```

```python
## a , b 입력 후 a>b로 출력하기
a, b = map(int,input("두개의 수를 입력하시오").split())
if(a>=b):
    print(f"{a} {b}")
else:
    print(f"{b} {a}")

```

```python
# 두 사람의 나이차 
a = int(input("첫번째 나이를 입력하시오"))
b = int(input("두번째 나이를 입력하시오"))
if(a>=b):
    diff = a-b
else:
    diff = b-a
print(f"나이차이: {diff}")
```

```python
#입력된 숫자가 양수,음수,0 인지 출력
num = int(input())
if(num>0):
    print("양수")
elif(num<0):
    print("음수")
else:
    print("0")
```

```python
##(숙제) 입력된 정수의 따라 등급출력 
score = int(input("점수를 입력하시오"))
if(score >=90):
    grade = "A"
elif(score >= 80):
    grade = "B"
elif(score >= 70):
    grade = "C"
elif(score >= 60):
    grade = "D"
else:
    grade = "F"
print(f"등급은 {grade}입니다")
```

```python
##(숙제)나이에따른 놀이공원 입장료
age = int(input("나이를 입력하시오"))
entry = 5000
if(age<8): entry = entry - (entry*0.3)
elif(age>=60): entry = entry - (entry*0.2)
print(f"입장료는 {entry}원 입니다")
```

```python
## 숫자를 입력받아 양수의경우에만 짝수인지 홀수인지 출력
num = int(input("숫자를 입력하시오"))
if(num>0):
    if(num%2==0):
        print("짝수입니다")
    else:
        print("홀수입니다")
else:
    print("양수가 아닙니다")
```
